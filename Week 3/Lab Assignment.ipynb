```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("wbc.csv")

df.head()
```

```python
df2 = df.copy()
```

1. Separate usable variables from unusable variables.

```python
df2.drop(columns=['id', 'Unnamed: 32'], inplace=True)
df2.head()
```

2. Encode the "diagnosis" column.

```python
le = LabelEncoder()
df2['diagnosis'] = le.fit_transform(df2['diagnosis'])
df2.head()
```

3. Standardize all columns with numeric values.

```python
# Separate features (X) and target (y)
X = df2.drop(columns=["diagnosis"])
y = df2["diagnosis"]
print("Before standarization")
print(X.mean())
print(X.std())

# Standarize numeric values
std = StandardScaler()
X_scaled = std.fit_transform(X)

# Convert back to DataFrame for readability
X_scaled_df2 = pd.DataFrame(X_scaled, columns=X.columns)

print("\nAfter standarization")
print(X_scaled_df2.mean())
print(X_scaled_df2.std())
```

4. Perform feature selection using SelectKBest.

```python
selector = SelectKBest(score_func=f_classif, k=5)
X_new = selector.fit_transform(X_scaled, y)

# Get mask of selected features
mask = selector.get_support()
selected_features = X.columns[mask]
print("Top 5 Selected Features:", list(selected_features))
```

5. Perform the testing process using the Logistic Regression model as in Lab 1.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Pipeline: StandardScaler, SelectKBest, LogisticRegression
pipeline = Pipeline([
    ('scaler', std),
    ('selector', selector),
    ('model', LogisticRegression(max_iter=1000))
])

# Training
pipeline.fit(X_train, y_train)

# Predict
pred = pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))

# Best features from pipeline
selector = pipeline.named_steps['selector']
mask = selector.get_support()
best_features = X.columns[mask]
print("The best features based on SelectKBest:", best_features.tolist())
```
